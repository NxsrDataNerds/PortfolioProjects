```python
#import libraries

import pandas as pd
import numpy as np
import panel as pn

pn.extension('tabulator')

import hvplot.pandas
```

```python
#Read Data

df = pd.read_csv(r'C:\Users\nasru\Downloads\Compressed\co2-data-master\owid-co2-data.csv')
```

```python
#Minor Data Preprocessing
#Change NAs with 0 & create GDP per capita column

df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population'] != 0, df['gdp']/df['population'], 0)
```

```python
#Dataframe Pipeline Interactive

idf = df.interactive()
```

```python
#hvplot CO2 emission over time by continent
#Create Panel Widgets

year_slider = pn.widgets.IntSlider(name = 'Year Slider', start = 1750, end = 2020, step = 5, value = 1850)
year_slider
```

```python
#Radio buttons for CO2 measures

yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = 'Y-axis',
    options = ['co2', 'co2 per capita',],
    button_type = 'success'
)
```

```python
#Create Pipeline

continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_plot = co2_pipeline.hvplot( x = 'year', by = 'country', y = yaxis_co2, line_width = 3, title = 'CO2 Emission By Continent')
co2_plot
```

```python
#Table for CO2 Emission by continent

co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination = 'remote', page_size = 10, sizing_mode = 'stretch_width')
co2_table
```

```python
#Scatterplot = CO2 VS GDP

co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by = 'year')
    .reset_index(drop = True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x = 'gdp_per_capita', y = 'co2', by = 'country', size = 100, kind = "scatter", alpha = 1, legend = False, height = 500, width = 1000)

co2_vs_gdp_scatterplot
```

```python
#Bar Chart for CO2 sources by continent

yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['coal_co2', 'oil_co2', 'gas_co2'],
    button_type = 'success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by = 'year')
    .reset_index(drop = True)
)
```

```python
co2_source_bar_plot = co2_source_pipeline.hvplot(
    kind = 'bar',
    x = 'country',
    y = yaxis_co2_source,
    title = 'CO2 Sources By Continents'
)
```
